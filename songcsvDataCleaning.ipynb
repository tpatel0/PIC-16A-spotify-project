```python
import pandas as pd
url = "https://raw.githubusercontent.com/lindseymardona/msd-exploration/main/SongCSV.csv"
ms = pd.read_csv(url)
```

```python
#get an idea of the columns of the data set
ms.head()
```

**Step 1:**
Remove Columns we are not looking at

```python
ms = ms[["SongNumber", "AlbumName", "ArtistName", "Title", "Year"]]
ms.head()
```

**Step 2:**
Remove the "b"s, " ", and " at the beginning of strings
Remove the " ", and " at the end of strings

```python
def cleanStrings(col):
    '''
    Removes leading "b"s, """s, and "'"s from all the strings in
    a dataframe column, as well as trailing """s and "'"s. 
    Args: col, a column from a dataframe, represented as a list.
    Returns: col, the cleaned & modified column.
    '''
    toClean = ['\'', '\"'] # unwanted leading & trailing characters
    index = 0
    while index < len(col): # goes through each list item
        item = col[index]
        cleaned = False
        
        while cleaned == False: # checks beginning & end of string
            if not item: # some strings are empty after cleaning
                cleaned = True 
            elif item[0] in toClean or item[0] == 'b': # if first item is to be deleted charatcer 
                item = item[1:] # reassign item beginning from the letter after                
            elif item[-1] in toClean:                  # if kast item is to be deleted character
                item = item[:-1] #reassign item to al of the letters before
            else:
                cleaned = True
                
        col[index] = item
                
        index += 1
        
    return col
```

```python
stringCols = ['AlbumName','ArtistName', 'Title']  
for col in stringCols:
    ms[col].dropna() # drop any nan values
    col = cleanStrings(ms[col]) # call above function to clean the data
ms.head() #get a look at how the data is cleaned
```

**Step 3**:
Drop rows with alphabet letters that aren't in the English alphabet

```python
for item in stringCols:
    ms = ms[ms[item].str.contains("\\",regex=False) == False] # uses string function to drop any non alphabetic chars
```

```python
ms.to_csv('clean.csv') #save file
```
